**Import Necessary Libraries**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Generate Sample Data**

```python
# Generate sample data
np.random.seed(42)
ages = np.random.randint(18, 65, size=5000)
genders = np.random.choice(['Male', 'Female'], size=5000, p=[0.55, 0.45])
```

**Create a DataFrame**

```python
# Create a DataFrame
data = pd.DataFrame({'Age': ages, 'Gender': genders})
```

** Calculate Age-Wise Population**

```python
# Age-wise population
age_counts = data['Age'].value_counts().sort_index()
```

**Calculate Gender-Wise Population**

```python
# Gender-wise population
gender_counts = data['Gender'].value_counts()
```

** Plot Age-Wise Population**

```python
# Plotting age-wise population
plt.figure(figsize=(10, 6))
plt.bar(age_counts.index, age_counts.values, color='skyblue', edgecolor='black')
plt.title('Age-wise Population')
plt.xlabel('Age')
plt.ylabel('Population')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

**Plot Gender-Wise Population**

```python
# Plotting gender-wise population
plt.figure(figsize=(6, 4))
plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'pink'], edgecolor='black')
plt.title('Gender-wise Population')
plt.xlabel('Gender')
plt.ylabel('Population')
plt.show()
```
